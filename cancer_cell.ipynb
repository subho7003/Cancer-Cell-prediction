```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cancer_data = pd.read_csv("cancer_data.csv")
```

```python
cancer_data = cancer_data.drop(["id"], axis=1)
```

```python
cancer_data["diagnosis"] = cancer_data ["diagnosis"].map({"M": 1, "B": 0})
```

```python
cancer_data = cancer_data.drop(["Unnamed: 32"], axis=1)
```

```python
sns.heatmap(cancer_data.corr())
```

```python
x = cancer_data.drop(columns=['diagnosis'])
y = cancer_data['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

LogisticRegression_model = LogisticRegression()
LogisticRegression_model.fit(x_train, y_train)
```

```python
LogisticRegression_model.score(x_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
sns.heatmap(confusion_matrix(y_test, LogisticRegression_model.predict(x_test)),annot=True)
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test, LogisticRegression_model.predict(x_test))
print(report)
```
